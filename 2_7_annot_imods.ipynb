```python
import pandas as pd
import os, sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from util.genome import is_overlap
from util.gene import get_gene_bnum
from util.feature import get_feat_d_from_ID
    
pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
```

```python
all_muts_df = pd.read_pickle("./data/2_6_df.pkl")
display(all_muts_df.shape, all_muts_df.head())
```

```python
TU_objects_df = pd.read_pickle("./data/TU_objects_df.pkl") 
TU_objects_df["TU range"] = TU_objects_df.apply(lambda row: (row["TU_POSLEFT"], row["TU_POSRIGHT"]), axis=1)
TU_objects_df
```

```python
imod_TU_range_df = pd.read_pickle("./data/imod_TU_df.pkl")
imod_TU_range_df.head()
```

```python
def get_TU_ID_set(feat_ID):
    return set(TU_objects_df[TU_objects_df["TU_OBJECT_ID"] == feat_ID]["TRANSCRIPTION_UNIT_ID"])


def get_imod_name_set(TU_ID):
    return set(imod_TU_range_df[imod_TU_range_df["TRANSCRIPTION_UNIT_ID"] == TU_ID]["imodulon"])


# Used regulons annotation NB as a model since it was vetted with the AVA publication.
def get_imodulons(mut_row, imod_TU_range_df):
    annot_json = []
    for feat_d in mut_row["genomic features"]:
        if '/' not in feat_d["RegulonDB ID"]:
            for tu in get_TU_ID_set(feat_d["RegulonDB ID"]):
                for imod in get_imod_name_set(tu):
                    annot_json.append({"name": imod})
        else:
            imod_TU_hit_df = imod_TU_range_df.copy()
            imod_TU_hit_df["mutation hit"] = imod_TU_hit_df["TU range"].apply(is_overlap, args=[feat_d["range"]])
            imod_TU_hit_df = imod_TU_hit_df[imod_TU_hit_df["mutation hit"]]  # keeping only those entries with a hit.
            for _, imod_TU in imod_TU_hit_df.iterrows():
                annot_json.append({"name": imod_TU["imodulon"]})

    return annot_json

all_muts_df["imodulons"] = all_muts_df.apply(lambda r: get_imodulons(r, imod_TU_range_df), axis=1)
```

```python
TU_df = pd.read_csv("./data/RegulonDB10/transcription_unit.txt",
                    sep="\t", comment='#', header=None)
TU_df.columns = [
    "TRANSCRIPTION_UNIT_ID",
    "PROMOTER_ID",
    "TRANSCRIPTION_UNIT_NAME",
    "OPERON_ID",
    "TRANSCRIPTION_UNIT_NOTE",
    "TU_INTERNAL_COMMENT",
    "KEY_ID_ORG"]


operon_df = pd.read_csv("./data/RegulonDB10/operon.txt",
                        sep="\t", comment='#', header=None)
operon_df.columns = [
    "OPERON_ID",
    "OPERON_NAME",
    "FIRSTGENEPOSLEFT",
    "LASTGENEPOSRIGHT",
    "REGULATIONPOSLEFT",
    "REGULATIONPOSRIGHT",
    "OPERON_STRAND",
    "OPERON_INTERNAL_COMMENT",
    "KEY_ID_ORG"]
operon_df["range"] = operon_df.apply(lambda row: (
    row["REGULATIONPOSLEFT"], row["REGULATIONPOSRIGHT"]), axis=1)
```

```python
# TODO: There's currently a function with the same name within util.operon. Refactor and combine the two
def get_operon_name_set(TU_ID):
    op_name_set = set()
    op_IDs = TU_df[TU_df["TRANSCRIPTION_UNIT_ID"]==TU_ID]["OPERON_ID"].unique()
    for op_ID in op_IDs:
        op_name_set.add(operon_df[operon_df["OPERON_ID"]==op_ID].iloc[0]["OPERON_NAME"])  # Assuming that each operon only has one name
    return op_name_set


def get_imodulon_operon_links_d(mut_row):
    imod_op_link_d = dict()
    for gen_targ_ID, feat_ID_set in mut_row["genetic feature links"].items():  # ??? Why using genetic and not genomic features?
        for feat_ID in feat_ID_set:
            feat_d = get_feat_d_from_ID(feat_ID, mut_row)
            
            # This part only handles mutations that land within a genomic feature.
            if '/' not in feat_d["RegulonDB ID"]:
                operon_name_set = set()
                imod_name_set = set()
                TU_mut_hit_set = get_TU_ID_set(feat_d["RegulonDB ID"])
                for tu in TU_mut_hit_set:
                    imod_name_set |= get_imod_name_set(tu)
                    operon_name_set |= get_operon_name_set(tu)  # using a set because only want one operon per mutated feature TU.
                for imod_name in imod_name_set:
                    if imod_name not in imod_op_link_d.keys(): imod_op_link_d[imod_name] = []
                    imod_op_link_d[imod_name] += list(operon_name_set)

            else:
                TU_mut_hit_df = TU_objects_df.copy()
                TU_mut_hit_df["mutation hit"] = TU_mut_hit_df["TU range"].apply(is_overlap, args=[feat_d["range"]])
                TU_mut_hit_df = TU_mut_hit_df[TU_mut_hit_df["mutation hit"]]
                TU_mut_hit_set = set(TU_mut_hit_df["TRANSCRIPTION_UNIT_ID"])
                operon_name_set = set()
                imod_name_set = set()
                for tu in TU_mut_hit_set:
                    imod_name_set |= get_imod_name_set(tu)
                    # using a set because only want one operon per mutated feature TU.
                    operon_name_set |= get_operon_name_set(tu)
                for imod_name in imod_name_set:
                    if imod_name not in imod_op_link_d.keys(): imod_op_link_d[imod_name] = []
                    imod_op_link_d[imod_name] += list(operon_name_set)
                    
    return imod_op_link_d


all_muts_df["imodulon links"] = all_muts_df.apply(lambda r: get_imodulon_operon_links_d(r), axis=1)
all_muts_df.head()
```
