```python
conds_d = {
    "glycine(2mM)  L-Serine(varying concentration)  trace elements(X1)  Wolfe's vitamin solution.": 'Sp',
    'ΔsdaA ΔsdaB ΔtdcG ΔglyA': 'S',
    '37 celsius': 'T',
    'M9': "M",
    'glucose(2)': 'C',
    'NH4Cl(1)': 'N',
    'KH2PO4(3) Na2HPO4(6.8)': 'P',
    'MgSO4(0.24)': 'Su',
    'CaCl2(0.1)': 'Ca',
    '511145': 'O',
}

leg = 'Sp: glycine(2mM) & L-Serine(varying concentration) supplement\n' + \
    'S: Δ$\it{sdaA}$ Δ$\it{sdaB}$ Δ$\it{tdcG}$ Δ$\it{glyA}$ starting strain\n' + \
    'T: 37 Celsius\n' + \
    'M: M9 minimal media\n' + \
    'C: glucose(2g/L) carbon source\n' + \
    'N: NH4Cl(1g/L) nitrogen source\n' + \
    'P: KH2PO4(2g/L) phosphorus source\n' + \
    'O: $\it{E. coli}$ K-12 MG1655\n' + \
    'Ca: CaCl2(0.1g/L) calcium source\n' + \
    'Su: MgSO4(0.24g/L) sulfur source'
leg
```

```python
import numpy as np


df = odds_df.copy()
df = df.rename(columns=conds_d)
df = df.applymap(lambda v: np.log(v) if v > 0 else v)  # get log-odds
df
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 100
sns.set_context("paper")
sns.set_style("white")


# re-ordering columns
l = df.columns.tolist()
l.remove("S")
l.remove("Sp")
l = ['Sp', 'S'] + l
df = df[l]

plt.figure(
    figsize=(2.2, 1.75)
)
ax = sns.heatmap(
    df,
    linewidths=0.1,
#     cmap="coolwarm",
    cbar_kws={
#         "shrink": 0.8,
        "label": "assocation\nlog odds ratio"
    }
)

plt.ylabel('key mutated\nfeatures')
plt.xlabel('conditions')

# fix for seaborn/matplotlib bug
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values


bbox_props = dict(
    boxstyle="round",
    fc="w",
    ec="0.5",
#     alpha=0.9
)
ax.annotate(leg,
            xy=(1.7, -0.1),
            xycoords='axes fraction',
#             textcoords='offset points',
#             size=14,
            bbox=bbox_props
           )

ax.set_xticklabels(
    ax.get_xticklabels(),
#     rotation=45,
#     horizontalalignment='right',
    verticalalignment='center',
#     rotation_mode='anchor'
#     position='top'
)

# ax.xaxis.set_label_position('top')
ax.xaxis.set_ticks_position('top')
ax.tick_params(axis='y', which='both', length=0)
ax.tick_params(axis='x', which='both', length=2, top=False)


plt.savefig("./figs/f5b.svg", format="svg")
plt.show()
```
