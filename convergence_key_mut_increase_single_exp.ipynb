```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = 100
mut_df = pd.read_pickle("./data/SER_df.pkl")
display(len(mut_df))
```

```python
def is_converging(mut_row, feat_col):
    is_converging = False
    for feat_d in mut_row[feat_col]:
        if feat_d["mutation set count"] > 1:
            is_converging = True
    return is_converging

FEAT_COLS = [
    "genomic features",
    "genetic features",
    "operons",
    "pathways",
    "regulators",
    "COGs"
]
for feat_col in FEAT_COLS:
    mut_df[feat_col+" convergence"] = mut_df.apply(lambda r: is_converging(r, feat_col), axis=1)
mut_df.head()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 100
sns.set_context("talk")
sns.set_style("ticks")
import numpy as np


count_d = {
    "all mutations": len(mut_df),
    "genomic feature converging mutation": np.sum(mut_df["genomic features convergence"]),
    "genetic feature converging mutation": np.sum(mut_df["genetic features convergence"]),
    "pathway converging mutation": np.sum(mut_df["pathways convergence"]),
    "operon converging mutation": np.sum(mut_df["operons convergence"]),
    "regulator converging mutation": np.sum(mut_df["regulators convergence"]),
    "COG converging mutation": np.sum(mut_df["COGs convergence"]),
}
df = pd.DataFrame.from_dict(count_d, orient='index', columns=["mutation count"])
df = df.sort_values("mutation count")
ax = df.plot.barh(y='mutation count',
#                   figsize=(5,10),
                  color="#4C72B0",
                  width=1
                )
for i, v in enumerate(df["mutation count"]):
    ax.text(v, i - 0.15, str(v))
ax.set_title("Mutation subset size")
sns.despine(ax=ax, top=True, right=True)
ax.get_legend().remove()
```

```python
# The intention is to demonstrate the cumulative convergence FOR MUTATIONS up to a certain resolution of annotation.
# Therefore, even if current level doesn't show convergence, possibly to due not being able to map with current level,
# if previously level demonstrated convergence, then going to annotate as convergence.
def is_converging_up_to_annot(mut_row, curr_feat_col, prev_feat_col=None):
    is_converging = False
    if prev_feat_col:  # if previous annotation type is found to converge, continue considering the mutation as converging.
        is_converging = mut_row[prev_feat_col]
    if not is_converging:  # if the previous column was not found to converge, the upcoming column could be found to converge.
        for feat_d in mut_row[curr_feat_col]:
            if feat_d["mutation set count"] > 1:
                is_converging = True
                break
    return is_converging

mut_df["genomic feature converging mutation"] = mut_df.apply(lambda r: is_converging_up_to_annot(r, "genomic features"), axis=1)
mut_df["genetic feature converging mutation"] = mut_df.apply(lambda r: is_converging_up_to_annot(r, "genetic features", "genomic feature converging mutation"), axis=1)
mut_df["operon converging mutation"] = mut_df.apply(lambda r: is_converging_up_to_annot(r, "operons", "genetic feature converging mutation"), axis=1)
mut_df["regulator converging mutation"] = mut_df.apply(lambda r: is_converging_up_to_annot(r, "regulators", "operon converging mutation"), axis=1)
mut_df["COG converging mutation"] = mut_df.apply(lambda r: is_converging_up_to_annot(r, "COGs", "operon converging mutation"), axis=1)
mut_df["pathway converging mutation"] = mut_df.apply(lambda r: is_converging_up_to_annot(r, "pathways", "operon converging mutation"), axis=1)
mut_df.head()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 100
sns.set_context("talk")
sns.set_style("ticks")
import numpy as np


count_d = {
    "genomic feature converging mutation": np.sum(mut_df["genomic feature converging mutation"]),
    "genetic feature converging mutation": np.sum(mut_df["genetic feature converging mutation"]),
    "operon converging mutation": np.sum(mut_df["operon converging mutation"]),
    "regulator converging mutation": np.sum(mut_df["regulator converging mutation"]),
}
prop_d = {
    "genomic feature converging mutation": np.sum(mut_df["genomic feature converging mutation"]/len(mut_df)),
    "genetic feature converging mutation": np.sum(mut_df["genetic feature converging mutation"]/len(mut_df)),
    "operon converging mutation": np.sum(mut_df["operon converging mutation"]/len(mut_df)),
    "regulator converging mutation": np.sum(mut_df["regulator converging mutation"]/len(mut_df)),
}
df = pd.DataFrame.from_dict(count_d, orient='index', columns=["mutation count"])
df["mutation proportion"] = pd.DataFrame.from_dict(prop_d, orient='index', columns=["mutation proportion"])
display(df)
df = df.sort_values("mutation count", ascending=False)
ax = df.plot.barh(y='mutation count',
#                   figsize=(5,10),
                  color="#4C72B0",
                  width=1
                )
for i, v in enumerate(df["mutation count"]):
    p = df.iloc[i]["mutation proportion"]
    ax.text(v, i - 0.15, str(v)+' ('+str('{0:.1f}'.format(p*100))+'%)')
ax.set_title(mut_df['exp'].unique()[0] + " converging mutation set size up to")
sns.despine(ax=ax, top=True, right=True)
ax.get_legend().remove()
```

```python
# The intention is to demonstrate the cumulative significant convergence FOR MUTATIONS up to a certain resolution of annotation.
# Therefore, even if current level doesn't show convergence, possibly to due not being able to map with current level,
# if previously level demonstrated convergence, then going to annotate as convergence.
def is_significant_up_to_annot(mut_row, curr_feat_col, prev_feat_col=None):
    is_converging = False
    if prev_feat_col:  # if previous annotation type is found to converge, continue considering the mutation as converging.
        is_converging = mut_row[prev_feat_col]
    if not is_converging:  # if the previous column was not found to converge, the upcoming column could be found to converge.
        for feat_d in mut_row[curr_feat_col]:
            if feat_d["significant"]:
                is_converging = True
                break
    return is_converging


mut_df["genomic feature significant mutation"] = mut_df.apply(lambda r: is_significant_up_to_annot(r, "genomic features"), axis=1)
mut_df["genetic feature significant mutation"] = mut_df.apply(lambda r: is_significant_up_to_annot(r, "genetic features", "genomic feature significant mutation"), axis=1)
mut_df["operon significant mutation"] = mut_df.apply(lambda r: is_significant_up_to_annot(r, "operons", "genetic feature significant mutation"), axis=1)
mut_df["regulator significant mutation"] = mut_df.apply(lambda r: is_significant_up_to_annot(r, "regulators", "operon significant mutation"), axis=1)
mut_df["COG significant mutation"] = mut_df.apply(lambda r: is_significant_up_to_annot(r, "COGs", "operon significant mutation"), axis=1)
mut_df["pathway significant mutation"] = mut_df.apply(lambda r: is_significant_up_to_annot(r, "pathways", "operon significant mutation"), axis=1)
mut_df.head()
```
