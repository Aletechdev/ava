```python
import pandas as pd
```

```python
gene_COG_df = pd.read_csv("./data/cogs_ecoli_mg1655.csv", encoding = "ISO-8859-1")
gene_COG_df.head()
```

```python
def _get_COGs(bnum):
    return set(gene_COG_df[gene_COG_df["locus"]==bnum]["COG description"])

all_gene_df = pd.read_csv("./data/y_ome/S1 y-ome Genes.tsv", sep='\t', header=0)
all_gene_df["COG"] = all_gene_df.apply(lambda g: _get_COGs(g["locus_tag"]), axis=1)

# Filter for only meaningful COGs
all_gene_df["COG"] = all_gene_df.apply(lambda g: set() if g["COG"]=={"Function unknown"} else g["COG"], axis=1)
all_gene_df["COG"] = all_gene_df.apply(lambda g: set() if g["COG"]=={"General function prediction only"} else g["COG"], axis=1)

all_gene_df
```

```python
all_gene_df[all_gene_df["COG"]!=set()]
```

```python
len(all_gene_df[all_gene_df["COG"]!=set()])/len(all_gene_df)
```

```python
# Get y-ome genes
y_ome_gene_df = all_gene_df[all_gene_df.yome=="yes"].copy()
y_ome_gene_df
```

```python
# Just to check if all COGs being mapped are meaningful
y_ome_COG_set = set()
for _, g in y_ome_gene_df.iterrows():
    y_ome_COG_set |= g["COG"]
y_ome_COG_set
```

```python
# The below shouldn't result in any y-genes being mapped to "Function unknown" only
for _, g in y_ome_gene_df.iterrows():
    if "Function unknown" in g["COG"]:
        display(g["COG"])
```
