```python
import pandas as pd
```

```python
pathway_df = pd.read_csv("./data/pathways/511145.12.PATRIC.pathway.tab", sep="\t")
pathway_df.head()
```

```python
def _get_pathways(bnum):
    return set(pathway_df[pathway_df["refseq_locus_tag"]==bnum]["pathway_name"])

all_gene_df = pd.read_csv("./data/y_ome/S1 y-ome Genes.tsv", sep='\t', header=0)
all_gene_df["pathway"] = all_gene_df.apply(lambda g: _get_pathways(g["locus_tag"]), axis=1)

all_gene_df
```

```python
all_gene_df[all_gene_df["pathway"]!=set()]
```

```python
len(all_gene_df[all_gene_df["pathway"]!=set()])/len(all_gene_df)
```

```python
# Get y-ome genes
y_ome_gene_df = all_gene_df[all_gene_df.yome=="yes"].copy()
y_ome_gene_df
```
