```python
# The intention is to demonstrate the cumulative convergence FOR MUTATIONS up to a certain resolution of annotation.
# Therefore, even if current level doesn't show convergence, possibly to due not being able to map with current level,
# if previously level demonstrated convergence, then going to annotate as convergence.
def is_converging_up_to_annot(mut_row, curr_feat_col, prev_feat_col=None):
    is_converging = False
    if prev_feat_col:  # if previous annotation type is found to converge, continue considering the mutation as converging.
        is_converging = mut_row[prev_feat_col]
    if not is_converging:  # if the previous column was not found to converge, the upcoming column could be found to converge.
        for feat_d in mut_row[curr_feat_col]:
            if feat_d["mutation set count"] > 1:
                is_converging = True
                break
    return is_converging

all_mut_df["genomic feature converging mutation"] = all_mut_df.apply(lambda r: is_converging_up_to_annot(r, "genomic features"), axis=1)
all_mut_df["genetic feature converging mutation"] = all_mut_df.apply(lambda r: is_converging_up_to_annot(r, "genetic features", "genomic feature converging mutation"), axis=1)
all_mut_df["operon converging mutation"] = all_mut_df.apply(lambda r: is_converging_up_to_annot(r, "operons", "genetic feature converging mutation"), axis=1)
all_mut_df["regulator converging mutation"] = all_mut_df.apply(lambda r: is_converging_up_to_annot(r, "regulators", "operon converging mutation"), axis=1)
all_mut_df["COG converging mutation"] = all_mut_df.apply(lambda r: is_converging_up_to_annot(r, "COGs", "operon converging mutation"), axis=1)
all_mut_df["pathway converging mutation"] = all_mut_df.apply(lambda r: is_converging_up_to_annot(r, "pathways", "operon converging mutation"), axis=1)
all_mut_df.head()
```

```python
# The intention is to demonstrate the cumulative significant convergence FOR MUTATIONS up to a certain resolution of annotation.
# Therefore, even if current level doesn't show convergence, possibly to due not being able to map with current level,
# if previously level demonstrated convergence, then going to annotate as convergence.
def is_significant_up_to_annot(mut_row, curr_feat_col, prev_feat_col=None):
    is_converging = False
    if prev_feat_col:  # if previous annotation type is found to converge, continue considering the mutation as converging.
        is_converging = mut_row[prev_feat_col]
    if not is_converging:  # if the previous column was not found to converge, the upcoming column could be found to converge.
        for feat_d in mut_row[curr_feat_col]:
            if feat_d["significantly enriched"]:
                is_converging = True
                break
    return is_converging


all_mut_df["genomic feature significant mutation"] = all_mut_df.apply(lambda r: is_significant_up_to_annot(r, "genomic features"), axis=1)
all_mut_df["genetic feature significant mutation"] = all_mut_df.apply(lambda r: is_significant_up_to_annot(r, "genetic features", "genomic feature significant mutation"), axis=1)
all_mut_df["operon significant mutation"] = all_mut_df.apply(lambda r: is_significant_up_to_annot(r, "operons", "genetic feature significant mutation"), axis=1)
all_mut_df["regulator significant mutation"] = all_mut_df.apply(lambda r: is_significant_up_to_annot(r, "regulators", "operon significant mutation"), axis=1)
all_mut_df["COG significant mutation"] = all_mut_df.apply(lambda r: is_significant_up_to_annot(r, "COGs", "operon significant mutation"), axis=1)
all_mut_df["pathway significant mutation"] = all_mut_df.apply(lambda r: is_significant_up_to_annot(r, "pathways", "operon significant mutation"), axis=1)
all_mut_df.head()
```
