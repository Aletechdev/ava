```python
import pandas as pd
pd.options.display.max_columns = 100
```

```python
muts_df = pd.read_pickle("./data/4_7_df.pkl")
display(muts_df.shape, muts_df.head())
```

```python
sample_col_name = "exp ale flask isolate"
muts_df[sample_col_name] = muts_df.apply(lambda r: r.exp + " " + str(r.ale) + " " + str(r.flask) + " " + str(r.isolate), axis=1)
muts_df.head()
```

```python
muts_df.columns
```

```python
# trying to get sample count per condition

m = pd.DataFrame()
cond_cols = set(['taxonomy-id', 'strain-description', 'base-media', 'temperature', 'carbon-source', 'supplement'])
drop_cols = set(muts_df.columns) - cond_cols - set([sample_col_name])
for col in cond_cols:
    df = muts_df.drop(columns=drop_cols)
    df = df.drop_duplicates()
    for cond in df[col].unique():
        m = m.append({"condition":cond,
                      "sample count": len(df[df[col]==cond]),
                      "type": col},
                     ignore_index=True)

m = m.set_index("condition")
m = m[["type", "sample count"]]
m
```

```python
muts_df.columns
```

```python
feat_types = set()
for feat_json in muts_df["genomic features"]:
    for f in feat_json:
        feat_types.add(f["feature type"])
feat_types
```

```python
for feat in feat_types:
    m[feat] = 0
m["operons"] = 0
m["regulons"] = 0


for feat_json in muts_df["genomic features"]:
#     print(feat_json)
    for f in feat_json:
        for sc in f['significantly associated conditions']:
            m.at[sc, f["feature type"]] += 1

for feat_json in muts_df["operons"]:
    for f in feat_json:
        for sc in f['significantly associated conditions']:
            m.at[sc, "operons"] += 1
for feat_json in muts_df["regulators"]:
    for f in feat_json:
        for sc in f['significantly associated conditions']:
            m.at[sc, "regulons"] += 1
            
m = m.rename(columns={"unknown": "unknown intergenic"})  # for the sake of exporting to document
m
```
