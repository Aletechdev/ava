```python
import pandas as pd
all_AVA_muts = pd.read_pickle("./data/4_7_df.pkl")
all_AVA_muts
```

```python
# Get y-ome genes
gene_df = pd.read_csv("./data/y_ome/S1 y-ome Genes.tsv", sep='\t', header=0)
y_ome_gene_df = gene_df[gene_df.yome=="yes"]
y_ome_gene_df
```

```python
gene_synonym_df = pd.read_pickle("./data/gene_synonym_df.pkl")
from util.gene import get_gene_bnum
```

```python
y_ome_gene_df[y_ome_gene_df["locus_tag"]=='']
```

```python
# Get mutations to y-ome genes
y_ome_muts = pd.DataFrame()
for _, m in all_AVA_muts.iterrows():
    for f in m["genetic features"]:
        bnum = ''
        if f["feature type"] == "gene":
            bnum = get_gene_bnum(f["RegulonDB ID"], gene_synonym_df)
        if bnum in set(y_ome_gene_df["locus_tag"]):
            y_ome_muts = y_ome_muts.append(m, ignore_index=True)
            break

y_ome_muts.to_pickle("./data/y_ome_muts.pkl")
y_ome_muts
```

```python
tot_mut_gene_instances = list()  # not just unique genes
y_ome_mut_genes_instances = list()
for _, m in all_AVA_muts.iterrows():
    for f in m["genetic features"]:
        bnum = ''
        if f["feature type"] == "gene":
            bnum = get_gene_bnum(f["RegulonDB ID"], gene_synonym_df)
            tot_mut_gene_instances.append(bnum)
        if bnum in set(y_ome_gene_df["locus_tag"]):
            y_ome_mut_genes_instances.append(bnum)
            
display(len(y_ome_mut_genes_instances), len(tot_mut_gene_instances), len(y_ome_mut_genes_instances)/len(tot_mut_gene_instances))
```
