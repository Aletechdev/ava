```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 300
sns.set_context("paper")
sns.set_style("ticks")

df = pd.DataFrame()
df = df.append({"type":"DEL regulatory region","count":2},ignore_index=True)
df = df.append({"type":"disrupt MOB catalytic region","count":2},ignore_index=True)
df = df.append({"type":"frameshift AMP catalytic region","count":1},ignore_index=True)
df = df.append({"type":"frameshift DEL regulatory region","count":4},ignore_index=True)
df = df.append({"type":"SNP catalytic region","count":7},ignore_index=True)
df = df.append({"type":"SNP regulatory region","count":5},ignore_index=True)
df["proportion"] = df['count'].apply(lambda cnt: cnt/sum(df["count"]))
df = df.sort_values(by="count", ascending=True)
display(df)

ax = df.plot.barh(
    y="count",
    x="type",
    color="#4C72B0",
    figsize=(2,1.25),
    width=1
)

df = df.reset_index()  # Resetting index for the below annotations to work
for i, r in df.iterrows(): 
    s = str(int(r["count"])) +' ({0:.1f}%)'.format(r["proportion"] * 100)
    ax.text(r["count"], i-0.3,  s)
sns.despine(ax=ax, top=True, right=True)
ax.get_legend().remove()
ax.set_title("The proportion of mutation targets\nfor $\it{cyaA}$")
ax.set_ylabel('')
ax.set_xlabel("mutation type count")
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 300
sns.set_context("paper")
sns.set_style("white")

df = pd.DataFrame()
df = df.append({"type":"uncharacterized","count":14},ignore_index=True)
df = df.append({"type":"mobile element insertion","count":2},ignore_index=True)
df = df.append({"type":"frameshift","count":5},ignore_index=True)
df["proportion"] = df['count'].apply(lambda cnt: cnt/sum(df["count"]))
df = df.sort_values(by="count", ascending=True)
display(df)

ax = df.plot.barh(
    y="count",
    x="type",
    color="#4C72B0",
    figsize=(2,0.75),
    width=1
)

plt.tick_params(
    axis='x',          # changes apply to the x-axis
    bottom=False,      # ticks along the bottom edge are off
    labelbottom=False) # labels along the bottom edge are off

df = df.reset_index()  # Resetting index for the below annotations to work
for i, r in df.iterrows(): 
    s = str(int(r["count"])) +' ({0:.1f}%)'.format(r["proportion"] * 100)
    ax.text(r["count"], i-0.3,  s)
sns.despine(ax=ax, top=True, right=True, left=True, bottom=True)
ax.get_legend().remove()
ax.set_title("Mutation effects to $\it{cyaA}$")
ax.set_ylabel('')
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 300
sns.set_context("paper")
sns.set_style("ticks")

df = pd.DataFrame()
df = df.append({"type":"uncharacterized","count":14},ignore_index=True)
df = df.append({"type":"disruptive","count":7},ignore_index=True)
df["proportion"] = df['count'].apply(lambda cnt: cnt/sum(df["count"]))
df = df.sort_values(by="count", ascending=True)
display(df)

ax = df.plot.barh(
    y="count",
    x="type",
    color="#4C72B0",
    figsize=(2,0.5),
    width=1
)

df = df.reset_index()  # Resetting index for the below annotations to work
for i, r in df.iterrows(): 
    s = str(int(r["count"])) +' ({0:.1f}%)'.format(r["proportion"] * 100)
    ax.text(r["count"], i-0.3,  s)
sns.despine(ax=ax, top=True, right=True)
ax.get_legend().remove()
ax.set_title("Predicted mutation effects\nto $\it{cyaA}$")
ax.set_ylabel('')
ax.set_xlabel("mutation type count")
```
