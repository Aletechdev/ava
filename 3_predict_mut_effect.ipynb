```python
from util.mut import is_frameshift


STOP_CODON_L = ["TAG", "TAA", "TGA"]
# This could be used for genetic mutations, though that currently uses Breseq generated annotations as a shortcut.
def is_pseudogene_readthrough_SNP(mut_df_row):
    is_readthru_SNP = False
    regdb_pgene_row = genes_df[genes_df["GENE_NAME"]==mut_df_row["RegulonDB Gene"]]
    stop_codon_gene_end = "GENE_POSRIGHT"
    if regdb_pgene_row.iloc[0]["GENE_STRAND"] == "reverse":
        stop_codon_gene_end = "GENE_POSLEFT"
    rel_stop_codon_gene_pos_chng = abs(regdb_pgene_row.iloc[0][stop_codon_gene_end] - mut_df_row["Position"])
    if rel_stop_codon_gene_pos_chng <= 2:
        # RegulonDB gives the sequence of the gene from left to right no matter what the strand.
        stop_codon_nt_l = list(regdb_pgene_row.iloc[0]["GENE_SEQUENCE"][-3:])
        stop_codon_pos_chng = int(3 - rel_stop_codon_gene_pos_chng - 1)  # last -1 due to being an index
        stop_codon_nt_l[stop_codon_pos_chng] = mut_df_row["Sequence Change"][-1]
        if ''.join(stop_codon_nt_l) not in STOP_CODON_L:
            is_readthru_SNP = True
    return is_readthru_SNP

# Not include a test genes_df because this test also serves to show if RegulonDB has changed component positions.
stop_codon_readthru_test_mut_df_row = pd.Series({"RegulonDB Gene": "ilvG_1", "Position": 3951543, "Sequence Change": "T→C"})
assert is_pseudogene_readthrough_SNP(stop_codon_readthru_test_mut_df_row) == True
stop_codon_silent_test_mut_df_row = pd.Series({"RegulonDB Gene": "ybbD", "Position": 528900, "Sequence Change": "A→G"})  # Current stop codon is TAA
assert is_pseudogene_readthrough_SNP(stop_codon_silent_test_mut_df_row) == False
reverse_strand_test_mut_df_row = pd.Series({"RegulonDB Gene": "wbbL_1", "Position": 2102914, "Sequence Change": "T→G"})  # Current stop codon is TGA
assert is_pseudogene_readthrough_SNP(reverse_strand_test_mut_df_row) == True
test_df = pd.Series({"RegulonDB Gene": "yfaS_1", "Position": 2330404, "Sequence Change": "A→C"})  # Current stop codon is TGA
assert is_pseudogene_readthrough_SNP(test_df) == False


def is_pseudogene_repaired(mut_df_row):
    is_repaired = False
    # Making the assumption that "_1" pseudogene part will always be the first, and therefore contain the premature stop codon regardless of strand.
    if mut_df_row["pseudogene"] and "_1" in mut_df_row["RegulonDB Gene"]:  # has to be in the first part of the pseudogene
        if (mut_df_row["Mutation Type"] == "SNP" and is_pseudogene_readthrough_SNP(mut_df_row)) \
        or (mut_df_row["Mutation Type"] not in ["SNP", "MOB"] and is_frameshift(mut_df_row["mutation size"])):
            is_repaired = True
    if mut_df_row["RegulonDB Gene"] == "gatC" \
    and mut_df_row.Position in [2173361, 2173362, 2173363, 2173364]  \
    and mut_df_row["Sequence Change"] == "Δ2 bp":
        is_repaired = True
        
    return is_repaired


all_muts_df["repaired"] = all_muts_df.apply(is_pseudogene_repaired, axis=1)
all_muts_df[all_muts_df["repaired"]]
```
